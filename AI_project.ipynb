```
import numpy as np
import wave
import os 


def loaddata (path):
    f = wave.open(path, 'rb')
    frames = f.readframes(-1)

    #Array of integers of range [0,255]
    data = np.fromstring(frames, dtype='uint8')
    
    #Normalized bytes of wav
    arr  = data/255
    return arr
```

```
traindata = []
trainlb = []
folder = 'AI_project_1'
files = os.listdir(folder)
files
```

```
for filename in files:
    data = loaddata (folder+"//" +filename)
    traindata .append(data)
    trainlb.append(filename[0])
    print(filename)
```

```
# Capping of the voice list ( cropping voice lists) 
length =[]
for voice in traindata:
    length.append(len(voice))
traindata_cropped = []
for voice in traindata:
    voice = voice[0:2000000]
    traindata_cropped.append(voice)
print(traindata_cropped)
```

```
traindata_cropped= np.matrix(traindata_cropped)
```

```
trainlb = np.array(trainlb).reshape(-1,1)

trainlb.shape
```

```
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
trainlb = le1.fit_transform(trainlb).reshape(-1,1)
```

```
print(type(trainlb))
print(trainlb.shape)
print(type(traindata_cropped))
print(traindata_cropped.shape)
```

```
from sklearn.linear_model import LogisticRegression

from keras.layers import Dense
```

```
model = LogisticRegression(C=5.0,max_iter=10000000,n_jobs=5,
                          verbose=1,penalty='l2')
```

```
model.fit(traindata_cropped,trainlb)
```

```
model.score(traindata_cropped,trainlb)
```
